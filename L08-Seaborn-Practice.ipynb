## Task 3a Using `relplot`

Experiment with the `size`, `hue` and `style` semantics by applying them to another example dataset of your choice.

*You should produce three or more plots for this task.*

```python
# loading the mpg dataset from the seaborn package

diamonds = sns.load_dataset('diamonds')
diamonds.head(10)
```

```python
# Carat vs Price using color as hue
sns.relplot(x = 'carat', y = 'price', hue = 'color', data = diamonds, aspect = 2)
```

```python
# depth vs table using cut as style

sns.relplot(x = 'depth', y = 'table', hue = 'color', style = 'cut', data = diamonds, aspect = 2)
```

```python
# carat vs price using color as hue, price as size and cut as style
sns.relplot(x = 'carat', y = 'price', hue = 'color', size = 'price', style = 'cut', data = diamonds, aspect = 2)
```

## Task 4a: Explore built-in styles

Using a dataset of your choice, practice creating a plot for each of these different styles:

+ darkgrid
+ whitegrid
+ dark
+ white
+ ticks

```python
# darkgrid style
sns.set_style('darkgrid')
sns.relplot(x = 'depth', y = 'table', hue = 'color', style = 'cut', data = diamonds, aspect = 2)
```

```python
# whitegrid style
sns.set_style('whitegrid')
sns.relplot(x = 'carat', y = 'price', hue = 'color', size = 'price', style = 'cut', data = diamonds, aspect = 2)
```

```python
# dark style
sns.set_style('dark')
sns.relplot(x = 'carat', y = 'price', hue = 'color', data = diamonds, aspect = 2)
```

```python
# white style
sns.set_style('white')
sns.relplot(x = 'depth', y = 'table', hue = 'color', style = 'cut', data = diamonds, aspect = 2)
```

```python
# ticks
sns.set_style('ticks')
sns.relplot(x = 'depth', y = 'table', hue = 'color', style = 'cut', data = diamonds, aspect = 2)
```
